## 2. Tractament de dades
El primer pas és importar la BBDD

```python
def load_dataset(path):
    dataset = pd.read_csv(path, header=0, delimiter=',', names=["ID", "Game", "State", "Hours", "0"])
    return dataset
dataset = load_dataset('../BBDD/steam-200k.csv')
print("Dimensionalitat de la BBDD:", dataset.shape)
print("Nombre de valors no existents:")
print(dataset.isnull().sum())
```

El primer tractament és eliminar la darrera columna *0* ja que és una columna buida que la propia web de la BBDD no fa referència.

Després s'uneixen les columnes *State* i *Hours* en una sola. Això es fa ja que per un mateix usuari pot existir un joc duplicat que cotingui els dos estats de `State€{play, pucharse}` indicant les hores que ha jugat i que l'ha comprat respectivament i, donat que `Hours€{0.1<x}` per tant mai hi haurà un valor 0, podem representar els jocs únicament comprats col.locant un 0 en la columna *Hours* que fins ara hi havia un 1 i si exiteixen jocs jugats 1 hora. Després cal eliminar els jocs repetits per un mateix usuari que compleixin Hours=0 ja que si ha jugat, vol dir que l'ha comprat.

Ara comptem amb una BBDD de dimensions 200.000x3, en quant a files anem servits ja que es dona la suficient informació i no ens hauriem de preocupar d'aplicar mètodes d'extensió de la BBDD per obtenir valors suficients per entrenar i per testejar. El problema ve en el següent paràmetre, les columnes, que tenim únicament 3 atributs amb els que treballar, per tant cal generar-ne més.

Aqui apareix la idea de valorar cada joc en funció tant del preu com de les hores jugades. La idea és establir una relació entre el que t'ha costat un joc i les hores que li has dedicat, sent el millor cas un joc barato que t'ha donat moltes hores. Primer caldrà generar el preu de cada joc (*Price*), el qual ho fem de manera aleatòria amb preus d'entre 10€ i 80€ fent salts de 5€. Ara tenim el paràmetre *Value* que ve de `*Price*/*Hours*`, s'ha decidit com a threshold que un bon idicador de bona compra és tenir un Value=2€/h, si aquest valor augmenta estàs "perdent" diners, mentre que si baixa millora l'indicador.

Aquests càlculs tenen una característica especial: el limit per x és 0 ja que per molt que juguis el joc no t'ha sortit gratis (no es consideren gratuïts) i volem que ens retorni un valor entre [0, 2]. Aquestes característiques es poden aprofitar amb una funció logaritmica inversa `y=10^(x)*2`.

![](Imatges/2grafiques.png) 
La funció de <font color='violet'>l'esquerra</font> és la que s'extreu de la funció escrita adalt, mentre que la de <font color='red'>la dreta</font> prové de fer la inversa i resulta `y=-log(x/2)`.

![](Imatges/1grafica.png)
Com es pot observar, la funció ens classifica un Value(€/h) en un valor d'entre [0,2], però clar, representat en el gràfic resultent diferents valors per tant per fer-ho resulta en 2€/h->0.02 i 1€/h->0.2 per tant 0==[2,0.2], 1==[0.2, 0.02] i 2==[0.02,0]. Les categories equivalen a {0:"Worthless", 1:"Fair enough worth", 2:"So worth"}.



```python
def set_dataset(dataset):
    datasetfinal=pd.DataFrame()
    #Per a cada jugador, ja que entre jugadors si hi haura repetits
    for ID in dataset["ID"].unique():
        #Juntem les columnes State i Hours
        #Separem del dataset entre State==play
        datasetID=dataset.loc[dataset["ID"]==ID]
        played=datasetID.loc[datasetID["State"]=='play']
        played=played.drop(['State'], axis=1)
        
        #Separem del dataset entre State==pucharse i eliminem els jocs que ja siguin en played
        purchased=datasetID.loc[datasetID["State"]=='purchase']
        purchased=pd.merge(purchased["Game"],played["Game"], indicator=True, how='outer').query('_merge=="left_only"').drop(['_merge'],axis=1)
        purchased=purchased.assign(ID=ID)
        columns_titles = ["ID","Game"]
        purchased=purchased.reindex(columns=columns_titles)
        purchased=purchased.assign(Hours=0)
          
        #Juntem played i pucharsed
        datasetID=pd.concat([played, purchased])
    
        PriceList=[]
        ValueList=[]
        LogList=[]
        WorthList=[]
        threshold=[0.02, 0.2]#1€/h -> 0.2, 2€/hora -> 0.02
        
        #Afegim el preu, el value es Preu/Hores, i li assignem una categoria entre 0,1,2
        for index, row in datasetID.iterrows():
            price=preusJocs[row["Game"]]
            PriceList.append(price)
            
            if row["Hours"] == 0:
                value=999
            else:
                value=price/row["Hours"]
                
            ValueList.append(value)
            value=(10**-value)*2
            LogList.append(value)
            
            if value < threshold[0]:
                WorthList.append(0)
            elif value < threshold[1]:
                WorthList.append(1)
            else:
                WorthList.append(2)
                
                
        datasetID["Price"]=PriceList
        datasetID["Value"]=ValueList
        datasetID["Log"]=LogList
        datasetID["Worth"]=WorthList
        datasetfinal=pd.concat([datasetfinal, datasetID])
    return datasetfinal

dataset = dataset.drop(['0'], axis=1)
dataset = dataset.loc[:10000, :]

Valor={0:"Worthless", 1:"Fair enough worth", 2:"So worth"}

#Definim un preu fix per a cada joc
preusJocs = {}
for game in dataset["Game"].unique():
    preusJocs[game]=random.randrange(10, 80, 5)
    
dataset=set_dataset(dataset)
```

Ara tenim una base de dades més lògica amb la que començar a treballar, de mida 200.000x7. Tenim un atribut Game, i la resta numèrics. Com no utilitzarem Game per a la classificació no cal fer-hi cap tractament.

## 3. Anàlisi de dades
Un cop amb la BBDD definitiva podem començar a veure les relacions entre els atributs. Per a fer-ho utilitzarem el scatter i el heatmap per analitzar les correlacions entre les dades. Ens podem imganiar que existirà gran relació entre els atributs Hours, Price, Value, Log i Worth ja que els atributs s'han generat a partir d'els altres.

```python
def draw_scatterheatmap(dataset):
    rangeY=2
    for i in range(1,len(dataset.columns)-1):
        atribut1=dataset.columns[i]
        for y in range(rangeY,len(dataset.columns)):
            atribut2=dataset.columns[y]
            plt.figure()
            plt.scatter(dataset[atribut1], dataset[atribut2])
            plt.xlabel(atribut1)
            plt.savefig('../AnalisiDades/scatter ' + atribut1 + "-"+ atribut2)
        rangeY+=1
        
    correlacio = dataset.corr()
    plt.figure()
    sns.heatmap(correlacio, annot=True, linewidths=.5)
    plt.savefig('../AnalisiDades/Heatmap')
    plt.show()

draw_scatterheatmap(dataset)
```

*Disclaimer: Només es mostra el heatmap ja que d'scatter n'hi han 20 elements, es millor accedir a la carpeta on es guarden (AnalisiDades).*
Com s'ha comentat, en el heatmap es mostra que existeix certa relació entre Log, Worth i Hours però curiosament no existeix amb el Price quan és l'element principal per generar aquests atributs. Això serà bo quan classifiquem amb els models ja que no els posarem tant fàcil aquesta classificació si en principi no es mostra relació entre els atributs.

## 4. Mètodes utilitzats
Els models escollits són: la regressió logistica (algoritme lienal) i la màquina de vectors SVM (algoritme no linear). El primer funciona amb funcions logaritmiques que imiten el comportament del càlcul realizat per classificar cada joc i volem comprovar que tant capaç és el model de trobar una ràpida solució i enfrontar-lo amb un model de funcionament diferent.

```python
x=dataset[["Hours", "Price"]]
y=dataset[["Worth"]]
y=y.values.ravel()
partes=[0.1, 0.2, 0.5]
for parte in partes:
    x_t, x_v, y_t, y_v = train_test_split(x, y, train_size=parte)
    
    # Creem el regresor logístic
    logireg = LogisticRegression()
    
    # l'entrenem
    logireg.fit(x_t, y_t)
    print("Correct classification Logistic ",parte, "% of the data: ", logireg.score(x_v, y_v))
    
    # Creem la maquina de vectors
    svc = svm.SVC(probability=True)
    
    # l'entrenem
    svc.fit(x_t, y_t)
    probs = svc.predict(x_v)
    print("Correct classification SVM      ",parte, "% of the data: ", svc.score(x_v, y_v))
```

Com era d'esperar la classificació per part del regressor logístic es execlent inclus amb només un 10% de les dades utilitzades per a entrenar els models. Mentres que el SVM queda una mica més enradere però segueix tenint un comportament excepcional.

## 5. Anàlisi de resultats
### Corves PR i ROC
Per a descartar un cas d'overfitting, poc probable degut a la poca porció de les dades utilitzades per a entrenar, utilitzares les corves PR i ROC per analitzar el rendiment dels models.

```python
lista_modelos = [LogisticRegression(), svm.SVC(probability=True)]
text_modelos = ["Logistic Regression", "SVM"]
part=0.4
for a, model in enumerate(lista_modelos):
    x_t, x_v, y_t, y_v = train_test_split(x, y, train_size=part)
    model.fit(x_t, y_t)
    probs = model.predict_proba(x_v)

    precision = {}
    recall = {}
    average_precision = {}
    plt.figure()
    for i in range(len(Valor)):
        precision[i], recall[i], _ = precision_recall_curve(y_v == i, probs[:, i])
        average_precision[i] = average_precision_score(y_v == i, probs[:, i])
        #plt.figure()
        #label='Precision-recall curve of class {0} (area = {1:0.2f})' ''.format(i, average_precision[i])
        plt.plot(recall[i], precision[i])
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.legend(loc="upper right")
    plt.savefig("../AnalisiResultats/Precission Recall Curve " + text_modelos[a])
    plt.show()
    
    fpr = {}
    tpr = {}
    roc_auc = {}
    for i in range(len(Valor)):
        fpr[i], tpr[i], _ = roc_curve(y_v == i, probs[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Compute micro-average ROC curve and ROC area
    # Plot ROC curve
    plt.figure()
    for i in range(len(Valor)):
        plt.plot(fpr[i], tpr[i], label='ROC curve of class {0} (area = {1:0.2f})' ''.format(i, roc_auc[i]))
    plt.savefig("../AnalisiResultats/ROC " + text_modelos[a])
    plt.show()
```

El dos primer gràfics es corresponen a les corves PR i ROC respectivament del Regressor Logistic, i les dues segones son les PR i ROC del SVM. Com es pot observar el regressor logistic realment classifica de manera casi perfecte degut a que el mètode utilitzat per classificar l'atribut Worth es similar al que utiliza aquest model. Mentre que la SVM s'observa un empitjorament de la classificació però segueix sent realment bona. Comentar que la <font color='orange'>classe 2</font> i la <font color='green'>classe 1</font> mostren una devallada en la precissió de classificació ja que n'hi han molts menos. El que es pot llegir com els usuaris tenen molts jocs comprats sense jugar o que juguen molt poc.

### Mershgrid

```python
def make_meshgrid(x, y, h=1):
        """Create a mesh of points to plot in
        Parameters
        ----------
        x: data to base x-axis meshgrid on
        y: data to base y-axis meshgrid on
        h: stepsize for meshgrid, optional
        Returns
        -------
        xx, yy : ndarray
        """
        x_min, x_max = x.min() - 1, x.max() + 1
        y_min, y_max = y.min() - 1, y.max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                              np.arange(y_min, y_max, h))
        return xx, yy

def plot_contours(ax, clf, xx, yy, **params):
    """Plot the decision boundaries for a classifier.
    Parameters
    ----------
    ax: matplotlib axes object
    clf: a classifier
    xx: meshgrid ndarray
    yy: meshgrid ndarray
    params: dictionary of params to pass to contourf, optional
    """
    a = xx.ravel()
    b = yy.ravel()
    c = np.c_[a, b]
    d = clf.predict(c)
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out

def show_C_effect(X, y, C=1.0, gamma=0.7, degree=3):

    # we create an instance of SVM and fit out data. We do not scale our
    # data since we want to plot the support vectors
    # title for the plots
    titles = ('SVC with linear kernel',
              'LinearSVC (linear kernel)',
              'SVC with RBF kernel',
              'SVC with polynomial (degree 3) kernel')

    #C = 1.0  # SVM regularization parameter
    models = (svm.SVC(),
              svm.LinearSVC(C=C, max_iter=1000000),
              svm.SVC(kernel='rbf', gamma=gamma, C=C),
              svm.SVC(kernel='poly', degree=degree, gamma='auto', C=C))
    models = (clf.fit(X, y) for clf in models)

    plt.close('all')
    fig, sub = plt.subplots(2, 2, figsize=(14,9))
    plt.subplots_adjust(wspace=0.4, hspace=0.4)


    X0 = X.loc[:, "Hours"]
    X1 = X.loc[:, "Price"]
    xx, yy = make_meshgrid(X0, X1)
    for clf, title, ax in zip(models, titles, sub.flatten()):
        plot_contours(ax, clf, xx, yy,
                      cmap=plt.cm.coolwarm, alpha=0.8)
        ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
        ax.set_xlim(xx.min(), xx.max())
        ax.set_ylim(yy.min(), yy.max())
        ax.set_xlabel('Sepal length')
        ax.set_ylabel('Sepal width')
        ax.set_xticks(())
        ax.set_yticks(())
        ax.set_title(title)
    plt.savefig("../AnalisiResultats/Scatter Meshgrid 2")
    plt.show()

show_C_effect(x, y, C=10)
```

No es molt interesant observar el mershgrid en aquest cas degut a l'abundant color vermell pertanyent a la classe 0. Però es mostra ja que es poden concloure certs resultats comentats en el apartat següent. Comentar que s'ha provat amb diferents valors de C i el resultat no ha variat.
